**Q4)** After reading the above file data, perform following string operations by writing Python code:
1. In the 2nd line of the file data, **extract from 4th element to 12th element**
2. In the above extracted data, perform **string reversal** of all the 9 elements
3. **Print all 9 elements** that were reversed in the above line


```python
class MyFileClass:
    def read_my_file_function(self, file_loc):
        # Opening a data file from the CURRENT LOCATION OF THE CODE
        with open(file_loc, "r") as fileobject:
            data = fileobject.read()
            data = data.splitlines()
            fileobject.close()
        # In the 2nd line of the file data, extract from 4th element to 12th element
        spliced_string = data[1][3:12]
        print(f"Range of Strings in the 2nd line of the file data is: {spliced_string}")
        #Reversing String
        reversed_string = spliced_string[::-1]
        print(f"Revered String is: {reversed_string}")
read_my_file_object = MyFileClass()
read_my_file_object.read_my_file_function("./asm1-data.txt")
```
